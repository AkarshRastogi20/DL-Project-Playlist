```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
# load the data from th sklearn
breast_cancer_data=sklearn.datasets.load_breast_cancer()
print(breast_cancer_data)
```

```python
# loading the data into dataframe
df=pd.DataFrame(breast_cancer_data.data,columns=breast_cancer_data.feature_names)
df
```

```python
# adding target column in the dataframe
df['label']=breast_cancer_data.target
df
```

**Separate label and features**

```python
X=df.drop(columns='label',axis=1)
Y=df['label']
```

```python
print(X)
```

```python
print(Y)
```

**Split the data into training and testing**

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

**Standardized the data**

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()

X_train_std=scaler.fit_transform(X_train)
X_test_std=scaler.transform(X_test)
```

```python
print(X_train_std)
```

**Building the Neural Network**

```python
import tensorflow as tf
tf.random.set_seed(3) # whenever you will run your neural n/w it will give us same accuracy score.
from tensorflow import keras
```

```python
# Now we have to build input layer,hidden layer and output layer
model=keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)), # first layer(flatten layer converts data into single array),30 columns
    keras.layers.Dense(20,activation='relu'), # Second layer(hidden layer),20 is neurons in hidden layer
    keras.layers.Dense(2,activation='sigmoid') # Output layer,2 neurons at output layer(as per the classes 0 & 1)
])
```

```python
# compiling the neural network

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy', # using this as a loss because of target variable 0 & 1
              metrics=['accuracy'])
```

**Train the neural network**

```python
history=model.fit(X_train_std,Y_train,epochs=100,validation_split=0.1) # validation split will perform testing
# our major goal should be increase the accuracy and reduce the loss
```

**Visualizing accuracy and loss**

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train','test'],loc='upper left')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train','test'],loc='upper left')
plt.show()
```

**Accuracy of the model on test data**

```python
loss,accuracy=model.evaluate(X_test_std,Y_test)
print(accuracy)
```

**Prediction of Model**

```python
Y_pred=model.predict(X_test_std)
print(Y_pred.shape)
print(Y_pred[0])
```
