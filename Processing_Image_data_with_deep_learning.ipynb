**Resizing the image using Pillow library**

```python
# Resizing the image

from PIL import Image
```

```python
img=Image.open('/content/dog.jpg')
img_resized=img.resize((200,200))
img_resized.save('/content/dog_resized.jpg')
```

```python
img_res=mpimg.imread('/content/dog_resized.jpg')
img_res_plot=plt.imshow(img_resized)
plt.show()
```

```python
print(img_res.shape) # we cannot find the shape of image untill unless convert into numpy by imread
```

**Converting RGB(Colorful) image into Grayscale(black&white) image using OpenCV**

```python
import cv2
```

```python
img=cv2.imread('/content/dog.jpg')
img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
# we can use cv2.imshow() will display the image but it does not
# support in google colab. that's why we can use google.colab.patches import cv2_imshow()

from google.colab.patches import cv2_imshow
cv2_imshow(img_gray)
```

```python
img_gray.shape
```

```python
img_gray
```

```python
# save the gray image
cv2.imwrite('/content/dog_gray.jpg',img_gray)
```
